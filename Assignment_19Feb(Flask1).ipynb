### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a simple and lightweight web framework for building web applications using Python. Its advantages include easy to learn and use, flexibility, extensibility, good for prototyping, and a large community for support.Flask is based on the Werkzeug WSGI toolkit and the Jinja2 template engine.

The advantages of using Flask include:

##### Easy to Learn and Use:
Flask is designed to be simple and easy to understand. It has a small and simple core, and it is easy to extend with third-party libraries.

##### Lightweight:
Flask is a lightweight framework that doesn't require much setup or boilerplate code. This makes it easy to use for small projects and prototypes.

##### Flexible:
Flask is a flexible framework that allows you to choose your own database, templating engine, and other tools. It does not impose any particular way of doing things.

##### Extensible:
Flask can be easily extended with third-party libraries to add functionality to your application. There are many third-party libraries available for Flask that can help you quickly add features to your application.

##### Good for Prototyping:
Flask is great for building prototypes quickly and easily. With its lightweight nature and flexibility, you can quickly get up and running with a basic application and start testing your ideas.

##### Large Community: 
Flask has a large community of developers and users who contribute to its development and provide support. This means that you can find help and resources easily if you run into any issues.

### Q3. What is App routing in Flask? Why do we use app routes?

In Flask, routing refers to the process of matching a URL to a particular function that generates a response. App routing in Flask is the process of mapping a URL to a specific view function in your application.

Routes are defined using the @app.route() decorator in Flask. This decorator tells Flask what URL should trigger the associated function. 
for above eg:- @app.route("/hello") means URL  is associated with the hello() function, which returns a message hello.


We use app routes in Flask to map URLs to view functions. This allows us to create different pages and endpoints for our web application. By defining different routes, we can handle different types of requests and provide different responses based on the URL that was requested. This is a fundamental feature of any web framework and allows us to build complex web applications with ease.

### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

The url_for() function is used to build a URL to the specific function dynamically. The first argument is the name of the specified function, and then we can pass any number of keyword argument corresponding to the variable part of the URL.

This function is useful in the sense that we can avoid hard-coding the URLs into the templates by dynamically building them using this function.
