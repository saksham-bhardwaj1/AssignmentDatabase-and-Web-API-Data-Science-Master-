### Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases?

MongoDb is a document-oriented NoSQL database system that uses JSON format to store data. it is designed to handle unstuctured and semi-structured data and provide high performance, scalability, and availability.
MongoDB is widely used in modern web applications,content management system, and big data processing systems.

Non-relational databases, also known as NoSQL databases, are databases that do not use the traditional tabular relational database model. unlike relational databases, NoSQL databases do not enforce strict schema and support flexible data models, which makes them ideal for handling unstructured and semistructured data such as text, multimedia, and social media data.

###### MongoDB is preferred over SQL databases in scenarios where:

##### Flexibility is required in the data model:
MongoDB's document-oriented data model allows developers to store and retrieve data without needing to define a schema beforehand, making it easier to work with flexible and unstructured data.

##### Scaling requirements:
MongoDB's sharding and replication features allow for easy horizontal scaling and high availability, which is difficult to achieve in traditional SQL databases.

##### Speed is crucial:
MongoDB is optimized for performance, providing high-speed queries and indexing of large data sets.

##### Agile development:
MongoDB's dynamic schema allows developers to make changes to the data model easily, which is particularly helpful in agile development environments.

### Q2. State and Explain the features of MongoDB.

MongoDB is a popular NoSQL database system that offers a wide range of features to handle unstructured and semi-structured data. Here are some of the key features of MongoDB:

##### Flexible Data Model:
MongoDB uses a document-oriented data model, which allows developers to store and retrieve data without needing to define a schema beforehand. This makes it easier to work with flexible and unstructured data.

##### High Performance: 
MongoDB is designed to provide high performance and scalability for large datasets. It supports horizontal scaling, sharding, and replication to improve performance and availability.

##### Indexing:
MongoDB provides a rich set of indexing features to support fast and efficient querying of data. It supports various indexing types such as single field, compound, multi-key, and geospatial indexes.

##### Querying:
MongoDB supports rich query expressions and provides a powerful aggregation framework for data analysis.

##### ACID Compliance:
MongoDB supports ACID (Atomicity, Consistency, Isolation, Durability) transactions for data consistency and reliability.

##### GridFS:
MongoDB provides a file storage system called GridFS, which allows you to store and retrieve large files, such as images or videos, directly in the database.

##### Security: 
MongoDB provides robust security features such as authentication, authorization, encryption, and auditing to ensure data privacy and integrity.

##### JSON-based: 
MongoDB stores data in a binary representation of JSON called BSON, making it easy to work with data in a modern web application environment.

In summary, MongoDB offers a wide range of features to support flexible, scalable, and high-performance data storage and retrieval. It is a popular choice for modern web applications, content management systems, and big data processing systems.

### Q7. Explain why delete_one(), delete_many(), and drop() is used.

In MongoDB, the delete_one() and delete_many() methods are used to delete one or many documents that match a specified filter criteria from a collection. On the other hand, the drop() method is used to drop an entire collection from a database.

Here's a brief explanation of why these methods are used:

##### delete_one():
This method is used to delete a single document that matches a specified filter from a collection. It is useful when you want to delete a specific document or only the first document that matches a filter.

##### delete_many(): 
This method is used to delete all the documents that match a specified filter from a collection. It is useful when you want to delete multiple documents that match a filter.

##### drop(): 
This method is used to drop an entire collection from a database. It is useful when you want to remove all documents and indexes from a collection or when you no longer need the collection and want to free up disk space.`

```python
## eg of delete one
delone=assignment_coll.delete_one({'salary':70000})
print(delone.deleted_count)  
```
