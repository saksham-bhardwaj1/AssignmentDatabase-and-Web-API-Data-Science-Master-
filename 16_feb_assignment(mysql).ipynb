### Q1. What is a database? Differentiate between SQL and NoSQL databases.

A database is an organized collection of data stored in a computer system, which can be accessed and manipulated by users or applications. A database management system(DBMS) is used to manage the storage , retrieval, modification, and deletion of data.

There are two main types of databases: SQL(Structured Query Language) and NoSql(Not only sql).

SQL databases are relational databases that use a structured language to manipulate data. They are based on a table-based model, where data is stored in tables with predefined relationships between them. Some popular SQL database are MYSQL,Oracle, and SQL Server.


NoSQL databases, on the other hand, are non-relational databases that do not use a structured language to manipulate data. They use a document-based ,key value, or graph-based model to store data which makes them more flexibleand scable than SQL databases.
NoSQL databases are particularly useful for handling large amounts of unstructured or semi-structured data, such as social media data, sensor data, or log files. Some popular NoSQL databases are MongoDB, Cassandra, and Redis.


In summary, SQL databases are best suited for applications that require complex querying and analysis of structured data, while NoSQL databases are better suited for applications that require flexibility, scalability, and fast access to unstructured or semi-structured data.

### Q5. Explain Primary Key and Foreign Key.

##### Primary Key:- 
A primary key is a column or a set in a table that uniquely identifies each row or record in that table
A primary key cannot contain null values and must be unique for each row. Typically, primary keys are used to establish relationships between tables in a database.
 
##### Foreign Key:-
A foreign Key , on other hand, is a column or a set of columns in a table that refers to the primary key of another table.
A foreign ket establishes a relationship between two tables and ensures referential integrity between them

### The table with the foreign key is called the child table, and the table with the primary key is called the referenced or parent table.

### Q7. Give the order of execution of SQL clauses in an SQL query.

In an SQL query, the clauses are executed in the following order:

##### FROM: 
This clause specifies the table or tables from which the data will be retrieved. If multiple tables are specified, they can be joined using the JOIN clause.

##### JOIN:
This clause combines data from two or more tables based on a common column or columns. There are several types of joins, including INNER JOIN, LEFT JOIN, RIGHT JOIN, and FULL OUTER JOIN. The type of join used determines which rows from each table are included in the result set.

##### WHERE:
This clause filters the data based on a specified condition. The condition can be a simple comparison, such as column_name = value, or a more complex expression involving logical operators like AND, OR, and NOT.

##### GROUP BY:
This clause groups the data based on one or more columns. When used with an aggregate function, such as SUM, COUNT, or AVG, it allows you to calculate summary information about the groups.

##### HAVING:
This clause filters the grouped data based on a specified condition. It is similar to the WHERE clause, but operates on the grouped data rather than the individual rows.

##### SELECT:
This clause selects the columns to be included in the result set. It can also include expressions that calculate new columns based on the existing data.

##### DISTINCT: 
This clause removes duplicate rows from the result set. It can be used in conjunction with the SELECT clause to return only unique values.

##### ORDER BY:
This clause sorts the result set based on one or more columns. By default, it sorts in ascending order, but you can also specify descending order using the DESC keyword.

##### LIMIT:
This clause limits the number of rows returned by the query. It can be used to return a specific number of rows, or to implement pagination by returning a range of rows.






Note that some of these clauses are optional, and not all of them are necessary in every query. However, if they are used, they must be used in the order listed above. For example, if a query includes a GROUP BY clause, the SELECT clause must also include the same columns that are used in the GROUP BY clause.
